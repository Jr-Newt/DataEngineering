#### 3. Give the maximum, mean, median, minimum age of the average targeted customer.


```python
db.bank.aggregate([
    {
        $group: {
            _id: null,
            ageData: { $push: "$age" }
        }
    },
    {
        $project: {
            _id: 0,
						minAge:{$min: "$ageData"},
						maxAge: {$max: "$ageData"},
						avgAge: {$avg: "$ageData"},
            medianAge: {$median : {input: "$ageData", method: "approximate"}}
        }
    }
])
```
